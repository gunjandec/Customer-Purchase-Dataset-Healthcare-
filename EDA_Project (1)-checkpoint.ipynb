## **Customer Purchase Dataset(Healthcare)**

## Purpose
The primary purpose of this project is to analyze patient appointment data to understand factors influencing no-show rates for medical appointments. By examining various attributes associated with the appointments and patients, the analysis aims to identify patterns and trends that can help medical facilities reduce no-show rates and improve patient care and operational efficiencynt (No for attended, Yes for missed).

## Objectives
The objectives of this project are:

Demographic Analysis: To analyze the demographic characteristics of patients (such as age, gender, and neighborhood) and their relationship with no-show rates. Health and Medical Factors: To examine how health conditions (like hypertension, diabetes, and alcoholism) and medical factors (such as SMS reminders) affect appointment attendance. Appointment Scheduling: To study the impact of scheduling details (such as the day and time of appointment scheduling) on no-show rates. Insights and Recommendations: To provide actionable insights and recommendations to help reduce no-show rates and enhance patient appointment management.

## Data Used
The dataset used for this analysis includes various attributes related to patient appointments. Below are the key columns in the dataset:

PatientId : Unique identifier for each patient.

AppointmentID : Unique identifier for each appointment.

Gender : Gender of the patient (F for Female, M for Male).

ScheduledDay : The date and time when the appointment was scheduled.

AppointmentDay : The date of the actual appointment.

Age : Age of the patient.

Neighbourhood : The neighborhood where the patient resides.

Scholarship : Indicates whether the patient is enrolled in the welfare program (0 for No, 1 for Yes).

Hipertension : Indicates whether the patient has hypertension (0 for No, 1 for Yes).

Diabetes : Indicates whether the patient has diabetes (0 for No, 1 for Yes).

Alcoholism : Indicates whether the patient has a history of alcoholism (0 for No, 1 for Yes).

Handcap : Indicates whether the patient is handicapped (0 for No, 1, 2, 3, or 4 indicating severity).

SMS_received : Indicates whether the patient received an SMS reminder (0 for No, 1 for Yes).

No-show : Indicates whether the patient showed up for the appointment (No for attended, Yes for missed).

## 4. Insights and Findings

## Insights and Findings

1. **Gender Influence**: The analysis shows that both male and female patients have similar no-show rates, indicating that gender does not significantly influence appointment attendance.
2. **Age Influence**: Younger patients (particularly children and teenagers) and older patients (above 60) have higher no-show rates. Targeted interventions may be required for these age groups.
3. **Neighbourhood Influence**: Certain neighbourhoods have higher no-show rates. Medical facilities in these areas might need to adopt specific strategies to improve attendance.
4. **Chronic Conditions**: Patients with certain chronic conditions like diabetes and hypertension have varied no-show rates. Tailored reminder systems and patient education programs could be beneficial.
5. **SMS Reminders**: Patients who received SMS reminders have a lower no-show rate, highlighting the importance of reminder systems in reducing no-shows.
6. **Time Between Scheduling and Appointment**: Longer gaps between scheduling and appointment dates are associated with higher no-show rates. Scheduling practices may need to be adjusted to minimize this gap.

These insights can help medical facilities implement data-driven strategies to reduce no-show rates and improve patient care and operational efficiency.
